# Customer Segmentation and Analysis






Analyzing customer data to segment the customer base based on various criteria, we can identify distinct customer segments using clustering algorithms and suggest targeted marketing strategies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visualize the distribution of Age, Annual Income, and Spending Score using histograms
plt.figure(1,figsize=(15,6))
n=0
for x in['Age','Annual Income (k$)','Spending Score (1-100)']:
    n+=1
    plt.subplot(1,3,n)
    plt.subplots_adjust(hspace = 0.5, wspace=0.5)
    sns.distplot(df[x], bins=20)
    plt.title('Distplot of {}'.format(x))
plt.show()
```

```python
# Visualize the count of customers by gender using a count plot
plt.figure(figsize=(15,5))
sns.countplot(y='Gender',data= df)
plt.show()
```

```python
# Visualize the distribution of Age, Annual Income, and Spending Score by gender using violin plots
plt.figure(1,figsize=(15,7))
n=0
for cols in['Age','Annual Income (k$)','Spending Score (1-100)']:
    n+=1
    plt.subplot(1,3,n)
    sns.set(style='whitegrid')
    plt.subplots_adjust(hspace = 0.5, wspace=0.5)
    sns.violinplot(x=cols, y='Gender', data=df)
    plt.ylabel('Gender' if n == 1 else '')
    plt.title('Violin Plot')
plt.show()
```

```python
# Create age groups and visualize the number of customers in each group
age_18_25 = df.Age[(df.Age >= 18) & (df.Age <= 25)]
age_26_35 = df.Age[(df.Age >= 26) & (df.Age <= 35)]
age_36_45 = df.Age[(df.Age >= 36) & (df.Age <= 45)]
age_46_55 = df.Age[(df.Age >= 46) & (df.Age <= 55)]
age_55above = df.Age[(df.Age >= 56)]

agex= ['18-25','26-35','36-45','46-55','55+']
agey= [len(age_18_25.values),len(age_26_35.values),len(age_36_45.values),len(age_46_55.values),len(age_55above.values)]

# Bar plot for age groups
plt.figure(figsize=(15,6))
sns.barplot(x=agex, y=agey, palette='mako')
plt.title('Number of Customer and Ages')
plt.xlabel('Age')
plt.ylabel("Number of Customer")
plt.show()
```

```python
# Perform KMeans clustering for Age vs Spending Score and visualize the clusters
# Use the Elbow Method to determine the optimal number of clusters (K)

import warnings
from sklearn.cluster import KMeans
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning)

X1=df.loc[:,['Age','Spending Score (1-100)']].values
from sklearn.cluster import KMeans
wcss = []
for k in range (1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(X1)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11), wcss, linewidth=2, color="red", marker="8")
plt.xlabel("K Value")
plt.ylabel("WCSS")
plt.show()

```

```python
kmeans = KMeans(n_clusters=4)

label = kmeans.fit_predict(X1)
print(label)
```

```python
print(kmeans.cluster_centers_)
```

```python
plt.scatter(X1[:,0],X1[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color='black')
plt.title('Clusters of Customers')
plt.xlabel('Age')
plt.ylabel('Spending Score(1-100)')
plt.show()
```

```python
# Perform KMeans clustering for Annual Income vs Spending Score and visualize the clusters
# Use the Elbow Method to determine the optimal number of clusters (K)

import warnings
from sklearn.cluster import KMeans
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning)

X2=df.loc[:,['Annual Income (k$)','Spending Score (1-100)']].values
from sklearn.cluster import KMeans
wcss = []
for k in range (1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(X2)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11), wcss, linewidth=2, color="red", marker="8")
plt.xlabel("K Value")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5)
label = kmeans.fit_predict(X2)
print(label)
```

```python
print(kmeans.cluster_centers_)
```

```python
plt.scatter(X2[:,0],X2[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color='black')
plt.title('clusters of Customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score(1-100)')
plt.show()
```

```python
# Perform KMeans clustering for Age, Annual Income, and Spending Score and visualize the clusters
# Use the Elbow Method to determine the optimal number of clusters (K)

import warnings
from sklearn.cluster import KMeans
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning)

X3=df.iloc[:,1:]
wcss = []
for k in range (1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(X3)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11), wcss, linewidth=2, color="red", marker="8")
plt.xlabel("K Value")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5)
label = kmeans.fit_predict(X3)
print(label)
```

```python
print(kmeans.cluster_centers_)
```

```python
# Visualize the clusters in a 3D scatter plot based on Age, Annual Income, and Spending Score with cluster labels

clusters =kmeans.fit_predict(X3)
df['label'] = clusters

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df.Age[df.label ==0], df['Annual Income (k$)'][df.label==0],df['Spending Score (1-100)'][df.label==0], c='blue',s = 60)
ax.scatter(df.Age[df.label ==1], df['Annual Income (k$)'][df.label==1],df['Spending Score (1-100)'][df.label==1], c='red',s = 60)
ax.scatter(df.Age[df.label ==2], df['Annual Income (k$)'][df.label==2],df['Spending Score (1-100)'][df.label==2], c='green',s = 60)
ax.scatter(df.Age[df.label ==3], df['Annual Income (k$)'][df.label==3],df['Spending Score (1-100)'][df.label==3], c='orange',s = 60)
ax.scatter(df.Age[df.label ==4], df['Annual Income (k$)'][df.label==4],df['Spending Score (1-100)'][df.label==4], c='purple',s = 60)
ax.view_init(30,185)
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
ax.set_zlabel('Spending Score (1-100)')
plt.show()
```
